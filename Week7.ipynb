```python
#b3
def bubble_sort(arr):
  n=len(arr)
  for i in range(n):
    check=0
    for j in range (n-i-1):
      if(arr[j] > arr[j+1]):
        arr[j],arr[j+1]=arr[j+1],arr[j]
        check=1
    if(check==0):
      break
  return arr
def selection_sort(arr):
  n=len(arr)
  for i in range(n):
    min_id=i
    for j in range(i+1,n):
      if(arr[j]<arr[min_id]):
        min_id=j
    arr[i],arr[min_id]=arr[min_id],arr[i]
  return arr
arr=list(map(int,input().split()))
print(selection_sort(arr))
print(bubble_sort(arr))
```
