```python
#b4
s=(map(str,input().split()))
dic={}
for pair in s:
  key,value=pair.split(':',1)
  if key not in dic:
    dic[key]=[value]
  else:
    dic[key].append(value)
print(dic)
```

```python
#b5
s=map(int,input().split())
dic={
    'posititive':0,
    'negatives':0,
    'Zero':0
}
for i in s:
  if (i>0):
    dic['posititive']+=1
  elif(i<0):
    dic['negatives']+=1
  else:
    dic['Zero']+=1
print(dic)
```

```python
#b9
dic={}
def nhapchuoi():
  s=input()
  s=s.replace('{','')
  s=s.replace('}','')
  s=s.replace("'",'')
  if(s==''): return
  s=map(str,s.split(','))
  for i in s:
    key,value=i.split(':',1)
    value=int(value)
    if(key in dic):
     dic[key]+=value
    else:
      dic[key]=value
nhapchuoi()
nhapchuoi()
print(dic)
```

```python
#b10
ans=[]
lst=list(map(int,input().split()))
k=int(input())
for i in lst:
  if (k-i in lst):
    tpl=(max(k-i,i),min(k-i,i))
    tpl1=(min(k-i,i),max(k-i,i))
    if(tpl not in ans):
      ans.append(tpl)
      if(tpl!=tpl1):
        ans.append(tpl1)
ans.sort()
print (ans)
```
