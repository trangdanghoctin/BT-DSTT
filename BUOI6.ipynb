```python
#b9
dic={}
def nhapchuoi():
  s=input()
  s=s.replace('{','')
  s=s.replace('}','')
  s=s.replace("'",'')
  if(s==''): return
  s=map(str,s.split(','))
  for i in s:
    key,value=i.split(':',1)
    value=int(value)
    if(key in dic):
     dic[key]+=value
    else:
      dic[key]=value
nhapchuoi()
nhapchuoi()
print(dic)
```

```python
#b10
ans=[]
lst=list(map(int,input().split()))
k=int(input())
for i in lst:
  if (k-i in lst):
    tpl=(max(k-i,i),min(k-i,i))
    tpl1=(min(k-i,i),max(k-i,i))
    if(tpl not in ans):
      ans.append(tpl)
      if(tpl!=tpl1):
        ans.append(tpl1)
ans.sort()
print (ans)
```

```python
#b14
def nhapchuoi():
  s1=list(map(int,input().split()))
  s2=list(map(int,input().split()))
  lst=[]
  if(len(s1)==0):
    return lst
  if(len(s2)==0):
    return lst
  for i in range(0,min(len(s1),len(s2))):
    if(s1[i] in s2 and s1[i] not in lst):
      lst.append(s1[i])
    if(s2[i] in s1 and s2[i] not in lst):
      lst.append(s2[i])
  return lst
print(nhapchuoi())
```
