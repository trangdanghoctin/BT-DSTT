```python
#b3
class soamError(Exception):
  pass
def tongbp(n):
  sum=0
  while(n):
    sum+=(n%10)**2
    n//=10
  return sum
while(True):
  try:
    n=int(input())
    f=[0]*(10**6)
    if(n>0):
      c=1
      while(c):
        if(n==1):
          print("YES")
          c=0
        elif(f[n]):
          print("NO")
          c=0
        else:
          f[n]=1
          n=tongbp(n)
      break
    else:
      raise soamError
  except ValueError:
    print("phai nhap mot so")
  except soamError:
    print("nhap 1 so duong")
```

```python
#b5
try:
  lst=list(map(float,input().split()))
  new_lst=[]
  for i in lst:
    if(i%1):
      raise ValueError("KHONG CHUYEN DUOC VE INT")
      break
    new_lst.append(int(i))
except ValueError as e:
  if(str(e)=="KHONG CHUYEN DUOC VE INT"):
    print(e)
else:
  print(new_lst)
```

```python
#b10
try:
  sum=0
  nums=list(map(int,input().split()))
  if (len(nums)!=len(set(nums))):
    raise ValueError
  else:
    for i in nums:
      if(i<0): raise ValueError
      sum+=i
    print(sum)
except ValueError:
  print("mang khong hop le")


```
