```python
#b13
def tonguoc(a):
  sum1=0
  for i in range (2,int(a**(1/2))):
   if(a%i==0 and int(a/i)!=i):
    sum1+=i
    sum1+=a/i
   if(a%i==0 and int(a/i)==i):
     sum1+=i
  return int(sum1+1)
number=list(map(int,input().split()))
print ("true" if (tonguoc(number[0])==number[1] and tonguoc(number[1])==number[0]) else "false")
```

```python
#b26
def chatnhiphan(a,num,l,r):
  ans=0
  while(l<=r):
    mid=l+int((r-l)/2)
    if(num[mid]>a):
      r=mid
      ans=r
    elif(num[mid]<=a):
      l=mid
  print(num[ans],ans,r)
num=[]
num.append(1)
num.append(2)
i=2
while(i<=99):
  num.append(num[i-1]+num[i-2])
  i=i+1
a=int(input())
chatnhiphan(a,num,1,99)
```
