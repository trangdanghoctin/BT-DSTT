```python
#b1
def merge(a,b):
  i=0
  j=0
  lst=[]
  while (i<len(a) and j<len(b)):
    if(a[i]<b[j]):
      lst.append(a[i])
      i+=1
    elif(a[i]>b[j]):
      lst.append(b[j])
      j+=1
    else:
      lst.append(a[i])
      lst.append(b[j])
      i+=1
      j+=1
  while i<len(a):
    lst.append(a[i])
    i+=1
  while j<len(b):
    lst.append(a[i])
    j+=1
  return lst
a=list(map(int,input().split()))
b=list(map(int,input().split()))
lst=merge(a,b)
for i in lst:
  print(i,end=" ")
```

```python
#b3
s=input()
s=list(s)
minn=[10]*(len(s)+2)
pos=len(s)-1
tmp=0
minpos=[10]*(14)
s.append('9')
for i in range(len(s)-1,-1,-1):
  if(int(s[i])<minn[i+1]):
    minpos[int(s[i])]=i
    minn[i]=int(s[i])
  else: minn[i]=minn[i+1]
for i in range(len(s)-1):
  if(int(s[i])>minn[i+1]):
    tmp=int(s[i])
    s[i]=str(minn[i+1])
    for j in range(i+1,len(s)):
      if(int(s[j])==minn[i+1]):
        if(int(s[j+1])>tmp):
          s[j]=str(tmp)
          break
      s[minpos[minn[i+1]]]=str(tmp)
    break
s.pop()
for i in s:
  print(i,end='')
```

```python
#b4
s1=input()
s2=input()
n=len(s1)
m=len(s2)
dp=[[0]*(m+1) for _ in range(n+1)]
#dp[i][j]: liệu i kí tự đàu tiên của s1 có tạo được j kí tự đầu tiên s2 không?
dp[0][0]=1
for i in range(1,n+1):
  for j in range(0,m+1):
    if(s1[i-1].islower()):
      dp[i][j]=dp[i-1][j]
      if(j>0 and s1[i-1].upper()==s2[j-1]):
        dp[i][j]=max(dp[i][j],dp[i-1][j-1])
    else:
      if(j>0 and s1[i-1]==s2[j-1]):
       dp[i][j]=dp[i-1][j-1]
print(dp[n][m])
```

```python
#b10
houses=list(map(int,input().split()))
heaters=list(map(int,input().split()))
heaters.sort()
def tknp(l,r,val):
  ans=-1
  while(l<r):
     m=l+(r-l)//2
     if(heaters[m]>=val):
      r=m-1
      ans=m
     else:
      l=m+1
  return ans
ans=0
for i in houses:
  pos=tknp(0,len(heaters),i)
  if(pos==-1):
    ans=max(ans,i-heaters[len(heaters)-1])
  else:
    if(i>0):
      ans=max(ans,min(heaters[pos]-i,i-heaters[pos-1]))
    else:
      ans=max(ans,heaters[pos]-i)
print(ans)

```
