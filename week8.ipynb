```python
#b1
class Cylinder():
  def __init__(self,bk,cc):
    self.bk=bk
    self.cc=cc
  def dientich(self):
    return (f"{((self.bk)*(self.bk)*3.14*2+2*3.14*(self.bk)*(self.cc)):.2f}")
  def thetich(self):
    return (f"{((self.bk)*(self.bk)*3.14*(self.cc)):.2f}")
bk,cc=map(float,input().split())
htru=Cylinder(bk,cc)
print(htru.dientich())
print(htru.thetich())
```

```python
#b2
from datetime import datetime
class Date:
  def __init__(self,d,m,y):
    self.d=d
    self.m=m
    self.y=y
  def __nextday__(self):
    x=self.m
    if(self.d== 31 and (x==1 or x==3 or x==5 or x==7 or x==8 or x==10)):
      self.d=1
      self.m+=1
      return
    if(self.d== 31 and x==12):
      self.d=1
      self.m=1
      self.y+=1
      return
    if(self.d==30 and (x==4 or x==6 or x==9 or x==11)):
      self.d=1
      self.m+=1
      return
    if((self.d== 28 or self.d==29) and x==2 ):
      self.d=1
      self.m=3
      return
    self.d+=1
def check(d,m,y):
  try:
    datetime(year=y, month=m, day=d)
    return 1
  except ValueError:
    return 0
d,m,y=map(int,input().split("/"))
date=Date(d,m,y)
if(check(d,m,y)):
  date.__nextday__()
  print(f"{date.d}//{date.m}//{date.y}")
else:
  print("INVALID")

```

```python
#b3
class point():
  def __init__(self,x,y):
    self.ox=x
    self.oy=y
  def check(self):
    if(self.ox==0 and self.oy==0):
      print("M là gốc tọa độ")
    elif(self.ox==0):
      print("M nằm trên trục tung")
    elif(self.oy==0):
      print("M nằm trên trục hoành")
    print(f"khoảng cách từ M đến gốc tọa độ là:{(((self.ox)**2+(self.oy)**2))**(1/2):.2f}")
x,y=map(float,input().split())
p=point(x,y)
p.check()
```

```python
#b6
class rectangle():
  def __init__(self,w,h):
    self.w=w
    self.h=h
  def area(self):
    return (self.w*self.h)
  def perimeter(self):
    return 2*(self.w+self.h)
  def scale(self,k):
    self.w*=k
    self.h*=k
w,h,k=map(float,input().split())
rec=rectangle(w,h)
rec.scale(k)
print(f"{rec.area():.2f}",f"{rec.perimeter():.2f}")
```

```python
#b7
class fraction():
  def gcd(self,a,b):
    while(a*b!=0):
      a,b=b,a%b
    return max(a,b)
  def __init__ (self,a,b):
    self.a=a//self.gcd(a,b)
    self.b=b//self.gcd(a,b)
  def __add__(self,fr2): # a/b + c/d =ad+cb/ bd
    m=self.a *fr2.b +fr2.a*self.b
    t=self.b*fr2.b
    print (f"{m//self.gcd(m,t)}/{t//self.gcd(m,t)}")
  def __sub__(self,fr2): # a/b - c/d =ad-cb/ bd
    m=self.a *fr2.b -fr2.a*self.b
    t=self.b*fr2.b
    print (f"{m//self.gcd(m,t)}/{t//self.gcd(m,t)}")
  def __mul__(self,fr2):
    m=self.a*fr2.a
    t=self.b*fr2.b
    print (f"{m//self.gcd(m,t)}/{t//self.gcd(m,t)}")
  def __floordiv__(self,fr2):
    m=self.a*fr2.b
    t=self.b*fr2.a
    print (f"{m//self.gcd(m,t)}/{t//self.gcd(m,t)}")
a,b,st,c,d=map(str,input().split())
a,b,c,d=int(a),int(b),int(c),int(d)
f1=fraction(a,b)
f2=fraction(c,d)
if(st=='+'):
  f1+f2
if(st=='-'):
  f1-f2
if(st=='*'):
  f1*f2
if(st=='/'):
  f1//f2

```

```python
#b8
class Bankaccount():
 def __init__(self,owner,balance):
  self.owner=owner
  self.balance=balance
 def deposit(self,deposit):
  self.balance+=deposit
 def withdraw(self,withdraw):
  if(withdraw > self.balance): pass
  else:
    self.balance-=withdraw
name,bl=map(str,input().split())
bl=float(bl)
ac1=Bankaccount(name,bl)
n=int(input())
while(n):
  promt,val=map(str,input().split())
  val=float(val)
  if(promt=="DEPOSIT"):
    ac1.deposit(val)
  else:
    ac1.withdraw(val)
  n-=1
print(f"{ac1.balance:.2f}")
```

```python
#b9
class Student():
  list=[]
  def __init__(self,name):
    self.name=name
    self.grade=[]
  def add_score(self,sub,score):
    self.grade.append({sub,score})
    self.list.append(score)
  def avg(self):
    avg=0
    for i,j in self.grade:
      avg+=j
    avg/=len(self.grade)
    print("Excellent" if avg>=8 else("Good" if avg>=6.5 else "Poor"))
name=input()
st1=Student(name)
n=int(input())
while(n):
  sub,sc=map(str,input().split())
  st1.add_score(sub,float(sc))
  n-=1
st1.avg()
```

```python
#b10
class book():
  def __init__(self,title,author,year):
    self.title=title
    self.author=author
    self.year=year
class librabry():
  def __init__(self):
    self.lst=[]
  def add(self,book):
    self.lst.append(book)
  def count_by_author(self,author):
    cnt=0
    for i in self.lst:
      if(i.author ==author):
        cnt+=1
    print(cnt)
  def count_by_year(self,year):
    cnt=0
    for i in self.lst:
      if(i.year ==year):
        cnt+=1
    print(cnt)
n=int(input())
librabryy=librabry()
while(n):
  s=input()
  cmd,cont=s.split(" ",1)
  if(cmd=="ADD"):
    cont=cont.split(';')
    book1=book(cont[0],cont[1],cont[2])
    librabryy.add(book1)
  elif(cmd=="COUNT"):
    librabryy.count_by_author(cont)
  else:
    librabryy.count_by_year(cont)
  n-=1
```

```python
#b11
import numpy as np
class plane():
  def __init__(self,p1,p2,p3):
    self.p1=p1
    self.p2=p2
    self.p3=p3
    self.n=[]
  def phaptuyen(self):
    p12=[]
    p13=[]
    for i in range(3):
      p12.append(self.p2[i]-self.p1[i])
      p13.append(self.p3[i]-self.p1[i])
    self.n.append(p12[1]*p13[2]-p12[2]*p13[1])
    self.n.append(p12[2]*p13[0]-p12[0]*p13[2])
    self.n.append(p12[0]*p13[1]-p12[1]*p13[0])
  def goc(self,other):
    tu=0
    tong1=0
    tong2=0
    for i in range(3):
      tu+=self.n[i]*other.n[i]
      tong1+=(self.n[i])**2
      tong2+=(other.n[i])**2
    tu=tu*(-1 if tu<0 else 1)
    mau=(tong1*tong2)**(1/2)
    if tu==0:
      return 90
    cos=tu/mau
    return (f"{np.rad2deg(np.arccos(cos)):.2f}")

n=4
p=[]
while(n):
  s=input()
  s=list(map(float,s.split()))
  p.append(s)
  n-=1
plane1=plane(p[0],p[1],p[2])
plane2=plane(p[1],p[2],p[3])
plane1.phaptuyen()
plane2.phaptuyen()
print(plane1.goc(plane2))

```
