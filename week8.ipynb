```python
#b1
class Cylinder():
  def __init__(self,bk,cc):
    self.bk=bk
    self.cc=cc
  def dientich(self):
    return (f"{((self.bk)*(self.bk)*3.14*2+2*3.14*(self.bk)*(self.cc)):.2f}")
  def thetich(self):
    return (f"{((self.bk)*(self.bk)*3.14*(self.cc)):.2f}")
bk,cc=map(float,input().split())
htru=Cylinder(bk,cc)
print(htru.dientich())
print(htru.thetich())
```

```python
#b2
from datetime import datetime
class Date:
  def __init__(self,d,m,y):
    self.d=d
    self.m=m
    self.y=y
  def __nextday__(self):
    x=self.m
    if(self.d== 31 and (x==1 or x==3 or x==5 or x==7 or x==8 or x==10)):
      self.d=1
      self.m+=1
      return
    if(self.d== 31 and x==12):
      self.d=1
      self.m=1
      self.y+=1
      return
    if(self.d==30 and (x==4 or x==6 or x==9 or x==11)):
      self.d=1
      self.m+=1
      return
    if((self.d== 28 or self.d==29) and x==2 ):
      self.d=1
      self.m=3
      return
    self.d+=1
def check(d,m,y):
  try:
    datetime(year=y, month=m, day=d)
    return 1
  except ValueError:
    return 0
d,m,y=map(int,input().split("/"))
date=Date(d,m,y)
if(check(d,m,y)):
  date.__nextday__()
  print(f"{date.d}//{date.m}//{date.y}")
else:
  print("INVALID")

```

```python
#b3
class point():
  def __init__(self,x,y):
    self.ox=x
    self.oy=y
  def check(self):
    if(self.ox==0 and self.oy==0):
      print("M là gốc tọa độ")
    elif(self.ox==0):
      print("M nằm trên trục tung")
    elif(self.oy==0):
      print("M nằm trên trục hoành")
    print(f"khoảng cách từ M đến gốc tọa độ là:{(((self.ox)**2+(self.oy)**2))**(1/2):.2f}")
x,y=map(float,input().split())
p=point(x,y)
p.check()
```
